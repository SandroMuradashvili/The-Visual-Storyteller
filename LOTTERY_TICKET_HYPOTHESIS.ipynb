```python
# ============================================================================
# NEURAL ARCHITECTURE SEARCH - BUILD & TRAIN SMALLER MODELS
# Creates genuinely smaller architectures with fewer actual parameters
# Compares: Full Model vs Medium vs Small vs Tiny architectures
# ============================================================================

!pip install torch torchvision timm matplotlib tqdm -q

import os
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence
import torchvision.transforms as transforms
import timm
import pickle
import matplotlib.pyplot as plt
from tqdm import tqdm
from PIL import Image
import pandas as pd
import time

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

print("\n" + "="*80)
print("NEURAL ARCHITECTURE SEARCH - SMALLER MODEL VARIANTS")
print("="*80)
print("We will BUILD and TRAIN completely different architectures:")
print("1. Full Model (256-dim, 4 heads, 3 layers) - Baseline")
print("2. Medium Model (128-dim, 4 heads, 2 layers) - 50% smaller")
print("3. Small Model (64-dim, 2 heads, 2 layers) - 75% smaller")
print("4. Tiny Model (32-dim, 2 heads, 1 layer) - 90% smaller")
print("="*80)

!pip install -q gdown
import gdown

file_id = "1st_oepUulGKCn5C1hvWvYYm87ftlUr6j"
url = f"https://drive.google.com/uc?id={file_id}"
gdown.download(url, "caption_data.zip", quiet=False)
!unzip -q caption_data.zip

# ===== Vocabulary Class Definition =====
# Need to define this BEFORE loading pickle to avoid errors
class Vocabulary:
    def __init__(self, freq_threshold=5):
        self.freq_threshold = freq_threshold
        self.itos = {0: "<PAD>", 1: "<SOS>", 2: "<EOS>", 3: "<UNK>"}
        self.stoi = {"<PAD>": 0, "<SOS>": 1, "<EOS>": 2, "<UNK>": 3}
        self.idx = 4

    def build_vocabulary(self, captions):
        from collections import Counter
        frequencies = Counter()
        for caption in captions:
            tokens = self.tokenize(caption)
            frequencies.update(tokens)
        for word, count in frequencies.items():
            if count >= self.freq_threshold:
                self.stoi[word] = self.idx
                self.itos[self.idx] = word
                self.idx += 1
        return self

    def tokenize(self, text):
        return text.lower().replace(',', '').replace('.', '').split()

    def numericalize(self, text):
        tokens = self.tokenize(text)
        return [self.stoi.get(token, self.stoi["<UNK>"]) for token in tokens]

    def __len__(self):
        return len(self.stoi)

# ===== Load Data & Vocabulary =====
print("\n[SETUP] Loading data and vocabulary...")

# Try to load existing vocabulary
try:
    with open('vocabulary.pkl', 'rb') as f:
        vocab = pickle.load(f)

    # Verify it loaded correctly
    if len(vocab.stoi) == 0 or "<PAD>" not in vocab.stoi:
        raise ValueError("Vocabulary corrupted, rebuilding...")

    print(f"✓ Vocabulary loaded: {len(vocab.stoi)} tokens")

except (FileNotFoundError, ValueError, AttributeError) as e:
    print(f"⚠ Vocabulary loading failed ({e}), rebuilding from scratch...")

    # Rebuild vocabulary from captions
    captions_df = pd.read_csv('captions.txt')
    vocab = Vocabulary(freq_threshold=5)
    vocab.build_vocabulary(captions_df['caption'].tolist())

    # Save it
    with open('vocabulary.pkl', 'wb') as f:
        pickle.dump(vocab, f)

    print(f"✓ Vocabulary rebuilt: {len(vocab.stoi)} tokens")

# Load captions and splits
captions_df = pd.read_csv('captions.txt')

with open('data_splits.pkl', 'rb') as f:
    splits = pickle.load(f)

print(f"✓ Train: {len(splits['train'])} | Val: {len(splits['val'])}")
print(f"✓ Special tokens: PAD={vocab.stoi['<PAD>']}, SOS={vocab.stoi['<SOS>']}, EOS={vocab.stoi['<EOS>']}")

# ===== Dataset Setup =====
class FlickrDataset(Dataset):
    def __init__(self, images_list, captions_df, vocab, img_dir, transform=None):
        self.images_list = images_list
        self.captions_df = captions_df[captions_df['image'].isin(images_list)]
        self.vocab = vocab
        self.img_dir = img_dir
        self.transform = transform

    def __len__(self):
        return len(self.captions_df)

    def __getitem__(self, idx):
        caption = self.captions_df.iloc[idx]['caption']
        img_name = self.captions_df.iloc[idx]['image']
        img_path = os.path.join(self.img_dir, img_name)

        try:
            image = Image.open(img_path).convert('RGB')
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")
            # Return a blank image as fallback
            image = Image.new('RGB', (224, 224), color=(128, 128, 128))

        if self.transform:
            image = self.transform(image)

        # Ensure vocab has required tokens
        sos_token = self.vocab.stoi.get("<SOS>", 1)
        eos_token = self.vocab.stoi.get("<EOS>", 2)

        numericalized = [sos_token]
        numericalized += self.vocab.numericalize(caption)
        numericalized.append(eos_token)

        return image, torch.tensor(numericalized)

class CollateFunc:
    def __init__(self, pad_idx):
        self.pad_idx = pad_idx

    def __call__(self, batch):
        images = torch.cat([item[0].unsqueeze(0) for item in batch], dim=0)
        captions = pad_sequence([item[1] for item in batch], batch_first=True, padding_value=self.pad_idx)
        return images, captions

val_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Use subset for faster experiments (remove this for full training)
train_subset = splits['train'][:8000]  # Use more samples
val_subset = splits['val'][:800]

train_dataset = FlickrDataset(train_subset, captions_df, vocab, 'Images', val_transform)
val_dataset = FlickrDataset(val_subset, captions_df, vocab, 'Images', val_transform)

batch_size = 32
pad_idx = vocab.stoi.get("<PAD>", 0)

print(f"✓ Using PAD index: {pad_idx}")

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True,
                          collate_fn=CollateFunc(pad_idx), num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False,
                        collate_fn=CollateFunc(pad_idx), num_workers=2, pin_memory=True)

print(f"✓ Training samples: {len(train_dataset)}")
print(f"✓ Validation samples: {len(val_dataset)}")

# ===== CONFIGURABLE MODEL ARCHITECTURE =====
class ImageEncoder(nn.Module):
    def __init__(self, embed_dim=256, use_pretrained=True):
        super().__init__()
        self.backbone = timm.create_model('efficientnet_b0', pretrained=use_pretrained)
        self.backbone.classifier = nn.Identity()
        self.projection = nn.Linear(1280, embed_dim)
        self.dropout = nn.Dropout(0.3)

    def forward(self, images):
        features = self.backbone(images)
        features = self.projection(features)
        features = self.dropout(features)
        return features.unsqueeze(1)

class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=100):
        super().__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-np.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        self.register_buffer('pe', pe.unsqueeze(0))

    def forward(self, x):
        return x + self.pe[:, :x.size(1)]

class TransformerDecoder(nn.Module):
    def __init__(self, vocab_size, embed_dim=256, num_heads=4, num_layers=3, dropout=0.3):
        super().__init__()
        self.embed_dim = embed_dim
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.pos_encoding = PositionalEncoding(embed_dim)

        # Dynamically calculate feedforward dim
        feedforward_dim = embed_dim * 4

        decoder_layer = nn.TransformerDecoderLayer(
            d_model=embed_dim,
            nhead=num_heads,
            dim_feedforward=feedforward_dim,
            dropout=dropout,
            batch_first=True
        )
        self.transformer_decoder = nn.TransformerDecoder(decoder_layer, num_layers)
        self.fc_out = nn.Linear(embed_dim, vocab_size)
        self.dropout = nn.Dropout(dropout)

    def forward(self, captions, image_features, tgt_mask=None, tgt_padding_mask=None):
        embedded = self.embedding(captions) * np.sqrt(self.embed_dim)
        embedded = self.pos_encoding(embedded)
        embedded = self.dropout(embedded)
        output = self.transformer_decoder(tgt=embedded, memory=image_features,
                                         tgt_mask=tgt_mask, tgt_key_padding_mask=tgt_padding_mask)
        return self.fc_out(output)

class ImageCaptioningModel(nn.Module):
    def __init__(self, vocab_size, embed_dim=256, num_heads=4, num_layers=3, dropout=0.3):
        super().__init__()
        self.encoder = ImageEncoder(embed_dim)
        self.decoder = TransformerDecoder(vocab_size, embed_dim, num_heads, num_layers, dropout)
        self.config = {
            'vocab_size': vocab_size,
            'embed_dim': embed_dim,
            'num_heads': num_heads,
            'num_layers': num_layers,
            'dropout': dropout
        }

    def forward(self, images, captions, tgt_mask=None, tgt_padding_mask=None):
        image_features = self.encoder(images)
        return self.decoder(captions, image_features, tgt_mask, tgt_padding_mask)

# ===== Training Functions =====
def generate_square_subsequent_mask(sz):
    return torch.triu(torch.ones(sz, sz), diagonal=1).bool()

def count_parameters(model):
    total = sum(p.numel() for p in model.parameters())
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    return total, trainable

def train_epoch(model, loader, criterion, optimizer, device):
    model.train()
    total_loss = 0

    for images, captions in tqdm(loader, desc="Training", leave=False):
        images, captions = images.to(device), captions.to(device)
        tgt_input, tgt_output = captions[:, :-1], captions[:, 1:]

        tgt_seq_len = tgt_input.size(1)
        tgt_mask = generate_square_subsequent_mask(tgt_seq_len).to(device)
        tgt_padding_mask = (tgt_input == pad_idx)

        optimizer.zero_grad()
        output = model(images, tgt_input, tgt_mask=tgt_mask, tgt_padding_mask=tgt_padding_mask)
        loss = criterion(output.reshape(-1, output.size(-1)), tgt_output.reshape(-1))

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(loader)

def validate(model, loader, criterion, device):
    model.eval()
    total_loss = 0
    with torch.no_grad():
        for images, captions in tqdm(loader, desc="Validation", leave=False):
            images, captions = images.to(device), captions.to(device)
            tgt_input, tgt_output = captions[:, :-1], captions[:, 1:]

            tgt_seq_len = tgt_input.size(1)
            tgt_mask = generate_square_subsequent_mask(tgt_seq_len).to(device)
            tgt_padding_mask = (tgt_input == pad_idx)

            output = model(images, tgt_input, tgt_mask=tgt_mask, tgt_padding_mask=tgt_padding_mask)
            loss = criterion(output.reshape(-1, output.size(-1)), tgt_output.reshape(-1))
            total_loss += loss.item()

    return total_loss / len(loader)

# ===== ARCHITECTURE CONFIGURATIONS =====
vocab_size = len(vocab.stoi)

ARCHITECTURES = {
    'Full': {'embed_dim': 256, 'num_heads': 4, 'num_layers': 3, 'dropout': 0.3},
    'Medium': {'embed_dim': 128, 'num_heads': 4, 'num_layers': 2, 'dropout': 0.3},
    'Small': {'embed_dim': 64, 'num_heads': 2, 'num_layers': 2, 'dropout': 0.3},
    'Tiny': {'embed_dim': 32, 'num_heads': 2, 'num_layers': 1, 'dropout': 0.3}
}

NUM_EPOCHS = 8  # Increase for better results
results = {}

# ===== TRAIN EACH ARCHITECTURE =====
for arch_name, config in ARCHITECTURES.items():
    print("\n" + "="*80)
    print(f"TRAINING: {arch_name} Model")
    print("="*80)
    print(f"Config: {config}")

    # Create model
    model = ImageCaptioningModel(
        vocab_size=vocab_size,
        **config
    ).to(device)

    # Count parameters
    total_params, trainable_params = count_parameters(model)
    print(f"✓ Total parameters: {total_params:,} ({total_params/1e6:.2f}M)")
    print(f"✓ Trainable parameters: {trainable_params:,}")

    # Setup training
    criterion = nn.CrossEntropyLoss(ignore_index=pad_idx)
    optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)

    # Training loop
    train_losses = []
    val_losses = []
    best_val_loss = float('inf')
    start_time = time.time()

    for epoch in range(NUM_EPOCHS):
        train_loss = train_epoch(model, train_loader, criterion, optimizer, device)
        val_loss = validate(model, val_loader, criterion, device)

        train_losses.append(train_loss)
        val_losses.append(val_loss)

        scheduler.step(val_loss)

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            # Save best model
            torch.save({
                'model_state_dict': model.state_dict(),
                'config': config,
                'vocab_size': vocab_size,
                'val_loss': val_loss
            }, f'model_{arch_name.lower()}.pth')

        print(f"Epoch {epoch+1}/{NUM_EPOCHS} - Train: {train_loss:.4f}, Val: {val_loss:.4f}")

    training_time = time.time() - start_time

    # Store results
    results[arch_name] = {
        'config': config,
        'total_params': total_params,
        'train_losses': train_losses,
        'val_losses': val_losses,
        'best_val_loss': best_val_loss,
        'final_val_loss': val_losses[-1],
        'training_time': training_time
    }

    print(f"\n✓ {arch_name} Model Complete!")
    print(f"  Best Val Loss: {best_val_loss:.4f}")
    print(f"  Training Time: {training_time/60:.1f} minutes")

# ===== RESULTS SUMMARY =====
print("\n" + "="*80)
print("ARCHITECTURE COMPARISON RESULTS")
print("="*80)

baseline = results['Full']

print(f"\n{'Model':<10} {'Params':<12} {'Size':<12} {'Best Val':<12} {'vs Full':<15} {'Time (min)':<12}")
print("-" * 80)

for name, res in results.items():
    size_reduction = (1 - res['total_params'] / baseline['total_params']) * 100
    perf_delta = ((res['best_val_loss'] - baseline['best_val_loss']) / baseline['best_val_loss']) * 100
    sign = "+" if perf_delta > 0 else ""

    print(f"{name:<10} "
          f"{res['total_params']/1e6:.2f}M{'':<6} "
          f"-{size_reduction:.0f}%{'':<8} "
          f"{res['best_val_loss']:.4f}{'':<6} "
          f"{sign}{perf_delta:.1f}%{'':<11} "
          f"{res['training_time']/60:.1f}")

# ===== VISUALIZATION =====
print("\nGenerating visualizations...")

fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# Plot 1: Training curves
ax1 = fig.add_subplot(gs[0, :2])
for name, res in results.items():
    ax1.plot(res['val_losses'], marker='o', label=name, linewidth=2, markersize=6)
ax1.set_xlabel('Epoch', fontsize=11)
ax1.set_ylabel('Validation Loss', fontsize=11)
ax1.set_title('Validation Loss: All Architectures', fontweight='bold', fontsize=13)
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)

# Plot 2: Model size comparison
ax2 = fig.add_subplot(gs[0, 2])
names = list(results.keys())
params = [results[n]['total_params']/1e6 for n in names]
colors = ['#2ecc71', '#3498db', '#f39c12', '#e74c3c']
bars = ax2.bar(names, params, color=colors, alpha=0.8)
ax2.set_ylabel('Parameters (Millions)', fontsize=11)
ax2.set_title('Model Size', fontweight='bold', fontsize=13)
for bar, val in zip(bars, params):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
            f'{val:.2f}M', ha='center', va='bottom', fontsize=9, fontweight='bold')

# Plot 3: Performance vs Size
ax3 = fig.add_subplot(gs[1, :2])
params_list = [results[n]['total_params']/1e6 for n in names]
losses = [results[n]['best_val_loss'] for n in names]
ax3.scatter(params_list, losses, s=200, c=colors, alpha=0.7, edgecolors='black', linewidth=2)
for i, name in enumerate(names):
    ax3.annotate(name, (params_list[i], losses[i]),
                xytext=(5, 5), textcoords='offset points', fontsize=10, fontweight='bold')
ax3.set_xlabel('Model Size (Million Parameters)', fontsize=11)
ax3.set_ylabel('Best Validation Loss', fontsize=11)
ax3.set_title('Performance vs Model Size (Pareto Frontier)', fontweight='bold', fontsize=13)
ax3.grid(True, alpha=0.3)
ax3.invert_yaxis()  # Lower loss is better

# Plot 4: Training time
ax4 = fig.add_subplot(gs[1, 2])
times = [results[n]['training_time']/60 for n in names]
bars = ax4.bar(names, times, color=colors, alpha=0.8)
ax4.set_ylabel('Training Time (minutes)', fontsize=11)
ax4.set_title('Training Time', fontweight='bold', fontsize=13)
for bar, val in zip(bars, times):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height,
            f'{val:.1f}m', ha='center', va='bottom', fontsize=9, fontweight='bold')

# Plot 5: Performance degradation
ax5 = fig.add_subplot(gs[2, :2])
size_reductions = [(1 - results[n]['total_params']/baseline['total_params'])*100 for n in names[1:]]
perf_degradations = [((results[n]['best_val_loss']-baseline['best_val_loss'])/baseline['best_val_loss'])*100 for n in names[1:]]
colors_subset = colors[1:]
bars = ax5.bar(names[1:], perf_degradations, color=colors_subset, alpha=0.8)
ax5.axhline(y=0, color='black', linestyle='-', linewidth=1)
ax5.axhline(y=5, color='orange', linestyle='--', linewidth=1, label='5% threshold')
ax5.axhline(y=10, color='red', linestyle='--', linewidth=1, label='10% threshold')
ax5.set_ylabel('Performance Degradation (%)', fontsize=11)
ax5.set_title('Loss Increase vs Full Model', fontweight='bold', fontsize=13)
ax5.legend(fontsize=9)
ax5.grid(True, alpha=0.3, axis='y')
for bar, val, size_red in zip(bars, perf_degradations, size_reductions):
    height = bar.get_height()
    ax5.text(bar.get_x() + bar.get_width()/2., height,
            f'{val:+.1f}%\n({size_red:.0f}% smaller)',
            ha='center', va='bottom' if val > 0 else 'top', fontsize=8, fontweight='bold')

# Plot 6: Efficiency score
ax6 = fig.add_subplot(gs[2, 2])
# Efficiency = Performance / Size (higher is better)
# Normalize: efficiency = (1 / loss) / params
efficiencies = [(1/results[n]['best_val_loss']) / (results[n]['total_params']/1e6) for n in names]
norm_eff = [e/max(efficiencies)*100 for e in efficiencies]
bars = ax6.bar(names, norm_eff, color=colors, alpha=0.8)
ax6.set_ylabel('Efficiency Score', fontsize=11)
ax6.set_title('Efficiency (Performance/Size)', fontweight='bold', fontsize=13)
ax6.set_ylim(0, 105)
for bar, val in zip(bars, norm_eff):
    height = bar.get_height()
    ax6.text(bar.get_x() + bar.get_width()/2., height,
            f'{val:.0f}', ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.savefig('architecture_comparison_results.png', dpi=150, bbox_inches='tight')
plt.show()

print("✓ Saved visualization to 'architecture_comparison_results.png'")

# ===== CONCLUSIONS =====
print("\n" + "="*80)
print("CONCLUSIONS & RECOMMENDATIONS")
print("="*80)

print("\nModel Recommendations:")
print("-" * 80)

for name, res in results.items():
    if name == 'Full':
        continue

    size_reduction = (1 - res['total_params'] / baseline['total_params']) * 100
    perf_delta = ((res['best_val_loss'] - baseline['best_val_loss']) / baseline['best_val_loss']) * 100

    print(f"\n{name} Model:")
    print(f"  • {size_reduction:.0f}% smaller ({res['total_params']/1e6:.2f}M params)")
    print(f"  • {perf_delta:+.1f}% performance change")

    if perf_delta < 5:
        print(f"  • ✓ RECOMMENDED: Excellent trade-off!")
    elif perf_delta < 10:
        print(f"  • ~ Good for resource-constrained deployment")
    else:
        print(f"  • ✗ Significant performance loss, not recommended")

print("\n" + "="*80)
print("Key Insights:")
print("="*80)
print("1. Smaller models train MUCH faster")
print("2. Moderate size reductions (30-50%) often maintain good performance")
print("3. Extreme compression (>75%) usually requires architectural innovations")
print("4. Choose based on your deployment constraints (mobile, server, etc.)")

print("\n✓ ARCHITECTURE SEARCH COMPLETE!")
print("✓ All models saved: model_full.pth, model_medium.pth, model_small.pth, model_tiny.pth")
print("="*80)
```
